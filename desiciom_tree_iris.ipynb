```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/Iris.csv')
df
```

```python
df.drop(columns='Id',inplace=True)
```

```python
x=df.iloc[:,:-1]
```

```python
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
dec=DecisionTreeClassifier(criterion='entropy')
dec.fit(x_train,y_train)
y_pred=dec.predict(x_test)
y_pred
```

```python
#model creation using decision tree
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(15,15))
tree.plot_tree(dec,feature_names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])
```

```python
#performence evaluation
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,accuracy_score,classification_report
result=confusion_matrix(y_test,y_pred)
DT_score=accuracy_score(y_test,y_pred)
report=classification_report(y_test,y_pred)
label=['Iris-versicolor',' Iris-setosa','Iris-virginica']
cmd=ConfusionMatrixDisplay(result,display_labels=label)
cmd.plot()
```

```python
# print('confusion matrix:',result)
result
```

```python
print('accuracy score of iris using decision tree classifier:',DT_score)
```

```python
#classification report
print(report)
```

```python
#Modelcreation using Randomforest classifier
from sklearn.ensemble import RandomForestClassifier
rm=RandomForestClassifier(n_estimators=10,criterion='entropy')
rm.fit(x_train,y_train)
y_pred=rm.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,accuracy_score,classification_report
result=confusion_matrix(y_test,y_pred)
RF_score=accuracy_score(y_test,y_pred)
report=classification_report(y_test,y_pred)
label=['Iris-versicolor',' Iris-setosa','Iris-virginica']
```

```python
print('Accuracy score of Iris using RandomForest classifier:',RF_score)
```



*   **Decision Tree Classifier Accuracy: 0.867**
*   **Random Forest Classifier Accuracy: 0.933**

>Based on the comparison, the Random Forest classifier achieved a higher accuracy score (0.933) compared to the Decision Tree classifier (0.867) for the Iris dataset. This suggests that the Random Forest model performed better in this scenario.
